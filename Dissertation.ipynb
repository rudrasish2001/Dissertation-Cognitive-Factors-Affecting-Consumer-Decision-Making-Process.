```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

def cronbach_alpha_score(itemscores):
    # number of items
    n_items = itemscores.shape[1]

    # variance of total scores
    variance_sum = np.var(np.sum(itemscores, axis=1), ddof=1)

    # sum of item variances
    sum_variance = np.sum(np.var(itemscores, axis=0, ddof=1))

    # Cronbach's alpha formula
    alpha = (n_items / (n_items - 1)) * (1 - sum_variance / variance_sum)

    return alpha

def calculate_alpha_if_item_removed(data, items):
    alphas = {}
    for item in items:
        print(f"Processing item: {item}")  # Debugging statement
        reduced_items = data.drop(columns=[item])
        print(f"Shape of data after removing {item}: {reduced_items.shape}")  # Debugging statement
        try:
            alpha = cronbach_alpha_score(reduced_items.to_numpy())
            alphas[item] = alpha
        except Exception as e:
            print(f"Error processing item {item}: {e}")  # Capture and print any error
    return alphas


alpha_if_item_removed = calculate_alpha_if_item_removed(data_encoded, all_questions_including_demographics)

# Print the results
for item, alpha in alpha_if_item_removed.items():
    print(f"Cronbach's Alpha if {item} is removed: {alpha}")

# Identify items where removing them increases the alpha
for item, alpha in alpha_if_item_removed.items():
    if alpha > overall_alpha_including_demographics_encoded:
        print(f"Removing {item} increases the overall Cronbach's Alpha to {alpha}")
```

```python
def convert_age_to_numeric(age_str):
    if '-' in age_str:
        # Convert ranges like "18-24" to the average of the range
        age_range = age_str.split('-')
        return (int(age_range[0]) + int(age_range[1])) / 2
    elif 'or older' in age_str:
        # Handle "65 or older" by taking the lower bound as the representative age
        return int(age_str.split(' ')[0])
    elif age_str.isdigit():
        return int(age_str)
    else:
        return None  # Handle non-numeric or unknown responses

# Apply conversion function to the age column
data['age_numeric'] = data['1. What is your age?'].apply(convert_age_to_numeric)

# Now define age categories using the cleaned numeric age data
bins = [18, 34, 54, np.inf]
labels = ['Younger', 'Middle-aged', 'Older']
data['age_category'] = pd.cut(data['age_numeric'], bins=bins, labels=labels, right=False)

# Proceed with the ANOVA as before
model = ols('Sustainable_Score ~ C(age_category)', data=data).fit()
anova_results = sm.stats.anova_lm(model, typ=2)

# Print the results
print(anova_results)
```
